```python
task = get_task("two_moons")
```

```python
simulator = task.get_simulator()
prior = task.get_prior_dist()
```

```python
thetas = prior.sample((1000,))
xs = simulator(thetas)
```

```python
loglikelihood = lambda x: task._likelihood(thetas,x)
```

```python
loglikelihood(torch.rand(1000,2, 100))
```
