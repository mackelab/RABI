```python
import svgutils.transform as sg
import sys 

from IPython.display import SVG,display

#create new SVG figure
fig = sg.SVGFigure(width="30.5cm", height="9cm")

# load matpotlib-generated figures
fig1 = sg.fromfile('pyloric_posterior_predictive.svg')
fig2 = sg.fromfile('pyloric_adv_post_marginals.svg')
fig3 = sg.fromfile("pyloric_posterior_predictive_rob.svg")
fig4 = sg.fromfile("pyloric_adv_post_marginals_rob.svg")

fig5 = sg.fromfile("robustness_metric.svg")
fig6 = sg.fromfile("coverages.svg")


scale = 1.33333333333
# get the plot objects
plot1 = fig1.getroot()
plot2 = fig2.getroot()
plot3 = fig3.getroot()
plot4 = fig4.getroot()
plot5 = fig5.getroot()
plot6 = fig6.getroot()

plot1.scale_xy(scale,scale)
plot2.scale_xy(scale,scale)
plot3.scale_xy(scale,scale)
plot4.scale_xy(scale,scale)
plot5.scale_xy(scale, scale)
plot6.scale_xy(scale,scale)

# plot1.moveto(130, 5)
# plot2.moveto(135, 80)
# plot3.moveto(330, 5)
# plot4.moveto(335,80)
# plot5.moveto(10, 3)
# plot6.moveto(10, 50)

scale_corrds = 2.16
plot1.moveto(1*scale_corrds, 0*scale_corrds)
plot2.moveto(12*scale_corrds, 93*scale_corrds)
plot3.moveto(200*scale_corrds, 0*scale_corrds)
plot4.moveto(207*scale_corrds,93*scale_corrds)
plot5.moveto(390*scale_corrds, 3*scale_corrds)
plot6.moveto(389*scale_corrds, 63*scale_corrds)

# add text labels
txt1 = sg.TextElement(6*scale_corrds,8*scale_corrds, "A", size=20, weight="bold", font="Arial")
txt2 = sg.TextElement(205*scale_corrds,8*scale_corrds, "B", size=20, weight="bold", font="Arial")
txt3 = sg.TextElement(390*scale_corrds,8*scale_corrds, "C", size=20, weight="bold", font="Arial")

# append plots and labels to figure
fig.append([plot1, plot2, plot3, plot4, plot5, plot6])
fig.append([txt1, txt2, txt3])

# save generated SVG files
fig.save("fig5_update.svg")


svg = SVG(filename='fig5_update.svg')
svg
```
