```python
import svgutils.transform as sg
import sys 

from IPython.display import SVG,display

#create new SVG figure
fig = sg.SVGFigure(width="30.5cm", height="16.5cm")

# load matpotlib-generated figures
fig1 = sg.fromfile('hh_posterior.svg')
fig2 = sg.fromfile('hh_predictive.svg')
fig3 = sg.fromfile("lotka_volterra_posterior.svg")
fig4 = sg.fromfile("lotka_volterra_predicitive.svg")
fig5 = sg.fromfile("sir_posterior.svg")
fig6 = sg.fromfile("sir_predictive.svg")
fig7 = sg.fromfile("vae_posterior.svg")
fig8 = sg.fromfile("vae_predictive.svg")
fig9 = sg.fromfile("spatial_sir_posterior.svg")
fig10 = sg.fromfile("spatial_sir_predictive.svg")

scale = 1.33333333
# get the plot objects
plot1 = fig1.getroot()
plot1.scale_xy(scale,scale)
plot2 = fig2.getroot()
plot2.scale_xy(scale,scale)
plot3 = fig3.getroot()
plot3.scale_xy(scale,scale)
plot4 = fig4.getroot()
plot4.scale_xy(scale,scale)
plot5 = fig5.getroot()
plot5.scale_xy(scale,scale)
plot6 = fig6.getroot()
plot6.scale_xy(scale,scale)
plot7 = fig7.getroot()
plot7.scale_xy(scale,scale)
plot8 = fig8.getroot()
plot8.scale_xy(scale,scale)
plot9 = fig9.getroot()
plot9.scale_xy(scale,scale)
plot10 = fig10.getroot()
plot10.scale_xy(scale,scale)
plot1.moveto(540*scale,133*scale)
plot2.moveto(510*scale, 8*scale)
plot3.moveto(40*scale,120*scale)
plot4.moveto(10*scale,5**scale)
plot5.moveto(126*scale,326*scale)
plot6.moveto(2.5*scale,323*scale)
plot7.moveto(310*scale,120*scale)
plot8.moveto(307*scale,0*scale)
plot9.moveto(385*scale,326*scale)
plot10.moveto(290*scale,326*scale)

# add text labels
txt1 = sg.TextElement(5*scale,15*scale, "A", size=20, weight="bold", font="Arial")
txt2 = sg.TextElement(291*scale,15*scale, "B", size=20, weight="bold", font="Arial")
txt3 = sg.TextElement(512*scale,15*scale, "C", size=20, weight="bold", font="Arial")
txt4 = sg.TextElement(5*scale,326*scale, "D", size=20, weight="bold", font="Arial")
txt5 = sg.TextElement(290*scale,326*scale, "E", size=20, weight="bold", font="Arial")

# append plots and labels to figure
fig.append([plot1, plot2, plot3, plot4, plot5,plot6, plot7, plot8, plot9, plot10])
fig.append([txt1, txt2, txt3, txt4, txt5])

# save generated SVG files
fig.save("fig2_appendix_trades.svg")


svg = SVG(filename='fig2_appendix_trades.svg')
svg
```
