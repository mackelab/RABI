```python
import svgutils.transform as sg
import sys 

from IPython.display import SVG,display

#create new SVG figure
fig = sg.SVGFigure(width="30.5cm", height="8.5cm")

# load matpotlib-generated figures
fig1 = sg.fromfile('sir_adversarial_thetas_eps_0.1.svg')
fig2 = sg.fromfile('sir_adversarial_thetas_eps_1.0.svg')
fig3 = sg.fromfile("sir_adversarial_thetas_eps_0.1_fKL.svg")
fig4 = sg.fromfile("sir_adversarial_thetas_eps_1.0_fKL.svg")


scale = 1.33333333
# get the plot objects
plot1 = fig1.getroot()
plot1.scale_xy(scale,scale)
plot2 = fig2.getroot()
plot2.scale_xy(scale,scale)
plot3 = fig3.getroot()
plot3.scale_xy(scale,scale)
plot4 = fig4.getroot()
plot4.scale_xy(scale,scale)

plot1.moveto(1*scale,0*scale)
plot2.moveto(220*scale, 0*scale)
plot3.moveto(445*scale,0*scale)
plot4.moveto(665*scale,0**scale)

# add text labels
txt1 = sg.TextElement(5*scale,15*scale, "A", size=20, weight="bold", font="Arial")
txt2 = sg.TextElement(230*scale,15*scale, "B", size=20, weight="bold", font="Arial")
txt3 = sg.TextElement(450*scale,15*scale, "C", size=20, weight="bold", font="Arial")
txt4 = sg.TextElement(660*scale,15*scale, "D", size=20, weight="bold", font="Arial")


# append plots and labels to figure
fig.append([plot1, plot2, plot3, plot4])
fig.append([txt1, txt2, txt3, txt4])

# save generated SVG files
fig.save("fig_thetas_appendix.svg")


svg = SVG(filename='fig_thetas_appendix.svg')
svg
```
